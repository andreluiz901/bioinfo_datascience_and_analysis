## Pre-processing data

### First, we want to mark qualitatively or data. In this case, we want to mark as 'active', 'intermediate' or 'inactive'. using the IC50 values to this.  

#### <1000 nM is considerate Active, >10000 nM is considerate Inactive and >1000 and <10000 its a Intermediate

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

As we need to make a list in the previus step, we can just iterate the few other columns we want and select all of them converting all lists to a new dataframe.

```python
#molecules id iteration
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)

#SMILES iteration
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)

#Standard_values iteration
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)

#combine all 4 lists in new dataframe
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
df3 #last call to confirm exact number of rows and columns
```

And, with new dataframe we want to save processed data in a new processed file

```python
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```

# Calculating Lipinski descriptors

These descriptors are used to evaluate druglikeness of compounds. Its based on pharmacokinetic profile: Absorption, Distribution, Metabolism and Excretion (ADME). 

The Lipinski's Rule stated the following:

    Molecular weight < 500 Dalton
    Octanol-water partition coefficient (LogP) < 5
    Hydrogen bond donors < 5
    Hydrogen bond acceptors < 10

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

Calculate descriptors

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df.canonical_smiles)
df_lipinski
```

Now, we combine the 2 dataframes: the preprocessed data and the calculated data.

```python
df_combined = pd.concat([df,df_lipinski], axis=1)
df_combined
```
